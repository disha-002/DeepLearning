```python
import numpy as np
```

```python
X=np.linspace(0,10,50).reshape(50,1)
# X represents an input data having one feature and 50 samples of it
# this range was chosen because it ensures we can see both early changes and later saturation
#means reshaping it to 1 column and as many rows as possible
```

```python
noise = np.random.normal(0,0.2,size=(50,1))
y=np.log(X+1) + noise
# since this target has a curved shape it cannot fit well by a straight line.
```

no of intermediate values -3

we want more than one so that the model can fit the data properly , too few cannot bend enough

taking too many will increase the computation.

```python
W1=np.random.uniform(-1,1,size=(1,3))
b1=np.zeros((1,3))
# means 1 input layer , 1 hidden layer having 3 units , 1 output layer
# each neuron in the hidden layer is passed 1 weight and and 1 bias
# so , we have 1,3 meaning one weight for each of the neurons

W2=np.random.uniform(-1,1,size=(3,1))
b2=np.zeros((1,1))

# weights are used to give te imporatance of the feature and the bias used as offset
# weights affect the shape , biases shift the curve up or down
```

```python
def activation(z):
  return np.maximum(0,z)
```

```python
def activation_slope(z):
  return (z>0).astype(float)

  # without a slope a model does not have a direction in which the parameters are to be changed
```

```python
# forward_pass
z1= X @ W1 + b1
# computes the weighted sum of inputs and bias for the hidden layer
h = activation(z1)
# applies activation function to the weighted sum
y_hat= h@W2 + b2
# predicts the output
```

```python
error = y_hat - y # computes the error
```

```python
loss = np.mean(error ** 2)

# since we are taking the square of the error very small values will not affect much but large values will bring about more change
```

```python
dL_dy = 2*error / len(X)
```

```python
dL_dW2 = h.T @ dl_dy
dL_db2 = np.sum(dl_dy, axis=0, keepdims=True)

# loss is function of y_hat , y_hat is a function of w2

# these depend on hidden values because loss does not directly depend on the parameters so we need to apply chain rule
```

```python
dL_dh = dl_dy @ W2.T
dL_dz1 = dL_dh*activation_slope(z1)

# loss is a function of y_hat and y_hat is a function of h that is a function of z1 which is a function of w1
# so to calculate whether the loss will increase of decrease wrt w1 , we need to apply chain rule and calculate the resyt of the slopes
```

```python
dL_dW1 = X.T @ dL_dz1
dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)
```

```python
learning_rate = 0.01
W1-= learning_rate * dL_dW1
b1-= learning_rate * dL_db1
W2-= learning_rate * dL_dW2
b2-= learning_rate * dL_db2

# in perceptron we update by multiplying learning rate with error
# here we are multiplying with the gradient , which gives us the direction in which the parameter needs to be changed
# since we are taking negative gradient , so we are moving in the direction opposite to where the loss is increasing
```

```python
for epoch in range(1000):
  z1= X @ W1 + b1
  h = activation(z1)
  y_hat = h@W2 + b2

  error = y_hat - y
  loss = np.mean(error ** 2)

  dL_dy = 2 * error / len(X)
  dL_dW2 = h.T @ dL_dy
  dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)

  dL_dh = dL_dy @ W2.T
  dL_dz1 = dL_dh * activation_slope(z1)
  dL_dW1 = X.T @ dL_dz1
  dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)

  W1 -= learning_rate * dL_dW1
  b1 -= learning_rate * dL_db1
  W2 -= learning_rate * dL_dW2
  b2 -= learning_rate * dl_db2

  if epoch % 100 == 0:
    print(f"Epoch {epoch}, Loss: {loss}")

# we should not take very large learning rate because it will overshoot the minimum
# very slow learning rate will cause the convergence very slow
```
