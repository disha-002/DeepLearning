```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Roll no - 102303705
a=0
b=5
min_N=30+a
#rotate one class by 15*b=15*5=75
theta = np.deg2rad(75)
np.random.seed(28)
```

```python
class0 = np.random.randn(min_N,2)
class1 = np.random.randn(min_N,2)+np.array([2.5,2.5])

#sd=0.01*a = 0.01*0 = 0. So, noise is 0.
```

```python
mean_1=np.mean(class1,axis=0)
class1_centered=class1-mean_1
```

```python
R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

class1_rotated = class1_centered @ R.T
class1 = class1_rotated + mean_1
```

```python
plt.scatter(class0[:,0], class0[:,1], label="Class 0")
plt.scatter(class1[:,0], class1[:,1], label="Class 1")
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.legend()
plt.title("Dataset with Relative Rotation")
plt.show()
```

```python
#Perceptron
#Class Labels (class0,class1)=>(-1,1)
X=np.vstack((class0,class1))
y=np.hstack((-1*np.ones(30),np.ones(30)))
```

```python
w = np.zeros(2)
b = 0
lr = 0.01
```

```python
for epoch in range(100):
  errors = 0
  for j in range(len(X)):
    xi = X[j]
    yi = y[j]
    if yi * (np.dot(w, xi) + b) <= 0:
        w += lr * yi * xi
        b += lr * yi
        errors += 1

    if errors == 0:
        print(f"Converged at epoch {epoch}")
        break
```

```python
x_vals = np.linspace(-2, 4, 100)
y_vals = -(w[0] * x_vals + b) / w[1]

plt.scatter(class0[:,0], class0[:,1], label="Class 0")
plt.scatter(class1[:,0], class1[:,1], label="Class 1")
plt.plot(x_vals, y_vals, '--', label="Perceptron Decision Boundary")
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.legend()
plt.show()
```

```python
# Convert labels from {-1, +1} to {0, 1}
y_log = (y == 1).astype(int)
```

```python
from sklearn.linear_model import LogisticRegression

# Create model
log_reg = LogisticRegression()

# Train on same dataset
log_reg.fit(X, y_log)
```

```python
w_lr = log_reg.coef_[0]   # weights
b_lr = log_reg.intercept_[0]  # bias
```

```python
x_vals = np.linspace(-2, 4, 100)
y_vals_lr = -(w_lr[0] * x_vals + b_lr) / w_lr[1]

plt.scatter(class0[:,0], class0[:,1], label="Class 0")
plt.scatter(class1[:,0], class1[:,1], label="Class 1")
plt.plot(x_vals, y_vals_lr, 'r-', label="Logistic Regression Boundary")
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.legend()
plt.title("Logistic Regression Decision Boundary")
plt.show()
```
