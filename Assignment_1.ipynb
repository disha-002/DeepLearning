```python
import numpy as np

# training function
def train_perceptron(X, y, lr=0.1, epochs=10):

    w = np.zeros(2)
    b = 0

    for epoch in range(epochs):

        for i in range(len(X)):

            x = X[i]

            z = np.dot(w, x) + b
            y_hat = 1 if z >= 0 else 0

            error = y[i] - y_hat

            w = w + lr * error * x
            b = b + lr * error

    return w, b
```

```python
def predict(X, w, b):

    preds = []

    for x in X:

        z = np.dot(w, x) + b
        y_hat = 1 if z >= 0 else 0

        preds.append(y_hat)

    return preds
```

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y_and = np.array([0,0,0,1])
y_or = np.array([0,1,1,1])
y_nand = np.array([1,1,1,0])
y_nor = np.array([1,0,0,0])
y_xor = np.array([0,1,1,0])
```

```python
def run_gate(name, X, y):

    w, b = train_perceptron(X, y)

    preds = predict(X, w, b)

    print("Gate:", name)
    print("Final weights:", w)
    print("Final bias:", b)

    for i in range(len(X)):
        print(X[i], "->", preds[i])

    print()
```

```python
run_gate("AND", X, y_and)
run_gate("OR", X, y_or)
run_gate("NAND", X, y_nand)
run_gate("NOR", X, y_nor)
run_gate("XOR", X, y_xor)
```

If the learning rate is too small the perceptron learns slowly and if it is too large then the descision boundary may oscillate.

since weights and biases are updated on the basis of error , different data causes different updates and hence same code can be used used for learning all gates.
